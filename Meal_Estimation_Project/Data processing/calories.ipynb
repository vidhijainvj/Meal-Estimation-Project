```python
class Calorie:
    
    def __init__(self,df):
        self.df = df
        pass
    
    def Cals(self):
        url = 'https://www.calories.info/'
        data = requests.get(url)
        soup = BeautifulSoup(data.content,'html.parser')
        s = soup.findAll('div',{'class':'par-33'})
        l=[]
        calories = []
        for html in s:
            l = html.ul.text.split('\n')
            l = l[1:-1]
            calories = calories+l
        return calories
    
    def apply(self):
        Calories = {'Cals':self.Cals()}
        
        calorie = []
        tokens = []
        for i,j in zip(df.Name,df.description):
            N = [k.lower() for k in i.split()] 
            try:
                D = [k.lower() for k in j.split()]
            except:
                D = []
                
            D = N+D[2:]
            tokens.append(D)
            
        for value in tokens:
            f=1
            for key,values in Calories.items():
                for i in [k.lower() for k in values]:
                    if i in value and f==1:
                        calorie.append(key)
    
                        f=0
                    
            if f==1:
                calorie.append(np.nan)
                f=0  
                
        self.df['Calorie'] = calorie
        
ob = Calorie(df)
ob.apply()

print('missing values:',df.Calorie.isnull().sum())
                                                
```
